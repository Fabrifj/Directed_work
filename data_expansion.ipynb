```python
# Group by 'Codigo_vendedor' and aggregate both the count and sum of 'Total_venta'
total_sells_by_seller = df.groupby("Codigo_vendedor")['Total_venta'].agg(['count', 'sum'])

# Calculate the average sale amount by dividing the sum by the count
total_sells_by_seller['avg_by_sell'] = total_sells_by_seller['sum'] / total_sells_by_seller['count']

total_sells_by_seller
```

```python
total_sells_by_seller['sum'].plot(kind="bar", figsize=(10,6), title="Total sells by seller")
plt.xlabel("Codigo de vendedor")
plt.ylabel("Suma de ventas totales")
plt.show()
```

```python
total_sells_by_seller['avg_by_sell'].plot(kind="bar", figsize=(10,6), title="avg sells by seller")
plt.xlabel("Codigo de vendedor")
plt.ylabel("Promedio de ventas por vendedor")
plt.show()
```

```python
total_sells_by_day = df.groupby("Fecha")['Total_venta'].agg(['count', 'sum'])

total_sells_by_day['avg_by_day'] = total_sells_by_day.apply(lambda x : x["sum"]/x["count"] if x["count"]>0  else 0, axis=1 )

total_sells_by_day
```

```python
total_sells_by_day['avg_by_day'].plot(kind="line", figsize=(10,6), title="avg sells by day")
plt.xlabel("Codigo de vendedor")
plt.ylabel("Promedio de ventas por vendedor")
plt.show()
```

# Data Expand #

### Date expand 

Add day, month and year

```python
df["Dia_semana"] = df["Fecha"].dt.strftime('%A')
df["Semana"] = df["Fecha"].dt.strftime("%U")
df["Numero_mes"] = df["Fecha"].dt.month
df["Ano"] = df["Fecha"].dt.year
```

```python
month_dictionary = {1:"Enero",2:"Febrero",3:"Marzo",4:"April",5:"Mayo",6:"Junio",7:"Julio",8:"Agosto",9:"Septiembre",10:"Octubre",11:"Noviembre",12:"Diciembre"}
df["Mes"] = df["Numero_mes"].map(month_dictionary)
```

### Client categories

Create a categorie from volum 

```python
clients_group = df.groupby(["Codigo_cliente"]).agg(
    Total_purchase=("Codigo_venta", "count"),
    Average=('Total_venta', 'mean')
)
```

```python
clients_group.sort_values(["Total_purchase"])
```

```python
clients_group = clients_group.reset_index()
clients_group_test = clients_group[(clients_group["Codigo_cliente"]!="o") & (clients_group["Codigo_cliente"]!="O")]
clients_group_test.describe()
```

**For Frecuency....|For Average**         
SMALL 0-50........      |SMALL 0-50th        
MEDIUM 50-75....        |MEDIUM 50-75        
LARGE 75>........       |LARGE  75>

```python
bins = [0, 5, 11, float('inf')]
labels = ["OCACIONAL", "HABITUAL", "FRECUENTE"]
clients_group["Frecuencia_cliente"] = pd.cut(clients_group["Total_purchase"], bins=bins, labels=labels, right=False)
```

```python
bins = [0, 1.259709e+03, 4.928304e+03, float('inf')]
labels = ["PEQUENO", "MEDIANO", "GRANDE"]
clients_group["Volumen_cliente"] = pd.cut(clients_group["Average"], bins=bins, labels=labels, right=False)
```

```python
print(clients_group.groupby(["Frecuencia_cliente"]).count())
print(clients_group.groupby(["Volumen_cliente"]).count())
```

```python
clients_group = clients_group.drop(columns=["Total_purchase", "Average"])
clients_group
```

```python
final_df = pd.merge(df, clients_group, on='Codigo_cliente', how='left')
final_df
```
